```python
from google.cloud import speech_v1p1beta1
```

```python
%env GOOGLE_APPLICATION_CREDENTIALS=./credential.json
```

```python
client = speech_v1p1beta1.SpeechClient()
```

```python
with open('5M617.mp3', 'rb') as audio_file:
    content = audio_file.read()
```

```python
gcs_uri = 'gs://thestandard-podcast/5minutes/5M617.mp3'
```

```python
#audio = speech_v1p1beta1.RecognitionAudio(content=content)
audio = speech_v1p1beta1.RecognitionAudio(uri=gcs_uri)
```

```python
config = speech_v1p1beta1.RecognitionConfig(
    encoding=speech_v1p1beta1.RecognitionConfig.AudioEncoding.MP3,
    sample_rate_hertz=44100,
    language_code="th-TH",
)
```

```python
operation = client.long_running_recognize(
    request={"config": config, "audio": audio}
)
```

```python
response = operation.result()
```

```python
for i in range(len(response.results)):
    print(response.results[i].alternatives[0].transcript)
    print('---')
```
